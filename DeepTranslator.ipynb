```python
def translate_to_english(text):
  """Translates text to English using GoogleTranslator.

  Args:
    text: The text to translate.

  Returns:
    The translated text.
  """
  translator = GoogleTranslator(source='auto', target='en')
  translated_text = translator.translate(text)
  return translated_text
```

```python
text_to_translate = "Hola, mundo!"  # Example Spanish text
translated_text = translate_to_english(text_to_translate)
print(f"Original text: {text_to_translate}")
print(f"Translated text: {translated_text}")
```

```python
text_to_translate = "म तिमीलाई धेरै प्रेम गर्छु"  # Example Nepali text
translated_text = translate_to_english(text_to_translate)
print(f"Original text: {text_to_translate}")
print(f"Translated text: {translated_text}")
```

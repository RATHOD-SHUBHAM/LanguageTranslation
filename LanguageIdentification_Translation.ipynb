```python
import fasttext
pretrained_lang_model = "/content/lid218e.bin" # Path of model file
model = fasttext.load_model(pretrained_lang_model)
text = "म तिमीलाई धेरै प्रेम गर्छु"
predictions = model.predict(text, k=1)
```

```python
predictions
```

```python
src = predictions[0][0].split('__label__')[1]
```

```python
src
```

```python
!pip install transformers
```

```python
from transformers import pipeline
```

```python
model_checkpoint = "facebook/nllb-200-distilled-600M"
```

```python
translator = pipeline("translation", model=model_checkpoint, src_lang=src, tgt_lang='eng_Latn')
translator(text)
```

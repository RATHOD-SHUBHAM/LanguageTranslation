```python
from transformers import pipeline
```

[Facebook nllb Model](https://huggingface.co/facebook/nllb-200-distilled-600M)

```python
model_checkpoint = "facebook/nllb-200-distilled-600M"
```

[Language_code](https://github.com/facebookresearch/flores/tree/main/flores200)

```python
translator = pipeline("translation", model=model_checkpoint, src_lang='eng_Latn', tgt_lang='hin_Deva')
translator("How old are you?")
```

```python
translator = pipeline("translation", model=model_checkpoint, src_lang='eng_Latn', tgt_lang='kan_Knda')
result = translator("How old are you?")
result[0]['translation_text']
```

```python
translator = pipeline("translation", model=model_checkpoint, src_lang='eng_Latn', tgt_lang='kan_Knda')
result = translator("We get 6 paid leaves per year?")
result[0]['translation_text']
```

```python
translator = pipeline("translation", model=model_checkpoint, src_lang='eng_Latn', tgt_lang='hin_Deva')
result = translator("We get 6 paid leaves per year?")
result[0]['translation_text']
```
